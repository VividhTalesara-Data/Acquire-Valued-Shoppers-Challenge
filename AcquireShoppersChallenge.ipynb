```python
data_split = {'id_in_train':id_in_train,'id_in_test':id_in_test,'id_in_conbine':id_in_conbine, 'id_in_chunk':id_in_chunk}

df_split = pd.DataFrame(data_split)

df_split
```

```python
# Feature No.1,2,3

id = []

category_record = []

brand_record = []

company_record = []

for n in idlist:
    
    category_list = df_1[df_1['id']==n]['category'].values
    
    category_name = df_2[df_2['id']==n]['category'].values
    
    brand_list = df_1[df_1['id']==n]['brand'].values
    
    brand_name = df_2[df_2['id']==n]['brand'].values
    
    company_list = df_1[df_1['id']==n]['company'].values
    
    company_name = df_2[df_2['id']==n]['company'].values
    
    if category_name in category_list:
        
        category_record_new = 1
        
    else:
  
        category_record_new = 0
    
    if brand_name in brand_list:
        
        brand_record_new = 1
        
    else:
  
        brand_record_new = 0
    
    if company_name in company_list:
        
        company_record_new = 1
        
    else:
  
        company_record_new = 0
    
    
    id.append(n)
    
    category_record.append(category_record_new)
    
    brand_record.append(brand_record_new)
    
    company_record.append(company_record_new)
    
%time
       
```

```python
feature_123 = pd.DataFrame(list(zip(id, category_record,brand_record,company_record)),columns=['id','category_record','brand_record','company_record'])
```

```python
# Feature No.4,5,6


id = []

category_spent = []

brand_spent = []

company_spent= []

for n in idlist:
    
    df_temp = df_1[df_1['id']==n]
    
    category_name = df_2[df_2['id']==n].squeeze()['category']
    
    category_spent_new = df_temp[(df_temp['category'].astype(int)== category_name) & (df_temp['purchaseamount']>0)]['purchaseamount'].sum()
    
    brand_name = df_2[df_2['id']==n].squeeze()['brand']
  
    brand_spent_new = df_temp[(df_temp['brand'].astype(int)== brand_name) & (df_temp['purchaseamount']>0)]['purchaseamount'].sum()
    
    company_name = df_2[df_2['id']==n].squeeze()['company']
    
    company_spent_new = df_temp[(df_temp['company'].astype(int)== company_name) & (df_temp['purchaseamount']>0)]['purchaseamount'].sum()
    
    id.append(n)
    
    category_spent.append(category_spent_new)
    
    brand_spent.append(brand_spent_new)
    
    company_spent.append(company_spent_new)
    
%time
     
```

```python
feature_456 = pd.DataFrame(list(zip(id, category_spent,brand_spent,company_spent))
                           ,columns=['id','category_spent','brand_spent','company_spent'])
```
